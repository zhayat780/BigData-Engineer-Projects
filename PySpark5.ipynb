# **Add new Column**

```
ms=df.withColumn('Copy_Date',df['Date']);
ms.show()
```

# **Add Multiple Columns at a time**



```
ms=df.withColumn('Copy_Date',df['Date']).withColumn('Double_Open',df['Open']*2).withColumn('Half_High',df['High']/2);
ms.show()
```

# **Rename Column**

```
hs=ms.withColumnRenamed("Copy_Date","Repeated_Date");
hs.show()
```

## **Rename Multiple Columns at a time**


```
hs=ms.withColumnRenamed("Copy_Date","Repeated_Date").withColumnRenamed("Adj Close","Close");
hs.show()
```

```
ms.show(2)
```

# **You can do Addition , Subtraction , Multiplication, Division etc on Columns**

```
ls=ms.withColumn("Double_Open",ms['Open']-2);
ls.show()
```

# Working with **SQL**

```
ls.createOrReplaceTempView('Public');
```

```
lg=spark.sql("select * from Public where Date='03-01-2012'");
lg.show()
```

# **Filtering Data based on certain condition**

```
df.show(5)
```

```
df.filter(df['Close']<60).select(['Volume','Open','Close']).show()
```

**Filter using Multiple Condition -- & for and , | for or , ~ for not , if multiple condition present , put them inside ()**

```
df.filter((df['Close']<80) & ~(df['Open']>70)).select(['Volume','Open','Close']).show()
```

# **Equality Condition Checking**

```
rs=df.filter(df['Low']==59.869999);
rs.show()
```

# **Collect the data as row object to work more on it**

```
mg=rs.collect()#returns one list
```

```
ls=mg[0]#take the first element of the list
```

```
ls#it's a row object
```

```
ps=ls.asDict()#Convert to dictionary , so that we can extract data later
```

```
ps#print the dictionary
```

```
ps['Close']#Extract data from Dictionary as key value format
```

```
df.show()
```

# **DataFrame column to List**

```
ls=df.select('Low');
ls.show()
```

```
type(ls)
```

# **Convert the selected Column to List of Row objects**

```
ms=ls.collect()
```

# **Iterate in the list , convert each row to dictionary , take the value part , append in a list**

```
mgs=[];
for i in ms:
  ps=i.asDict();
  mgs.append(ps['Low']);
```

# **Now we can apply all the methods which are applicable on the list data structure to inspect the columns**

```
max(mgs)
```

# **Frequently used Aggregate Functions**

```
df.select(sum('Close').alias("Total Value")).show()
```

```
df.select(mean('Close').alias("Average Value")).show()
```

```
df.select(avg('Close').alias("Average Value")).show()
```

```
df.select(max('Close').alias("Maximum Value")).show()
```

```
df.select(min('Close').alias("Minimum Value")).show()
```

```
df.select(stddev('Close').alias("Standard Deviation")).show()
```

```
df.select(variance('Close').alias("Variance")).show()
```
