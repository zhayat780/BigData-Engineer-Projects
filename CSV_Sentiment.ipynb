```python
import pandas as pd
import boto3
```

```python
aws_access_key=""
aws_secret_key=""
```

```python
session = boto3.Session(aws_access_key_id=aws_access_key,
    aws_secret_access_key=aws_secret_key,
    region_name='us-east-1'
)
comprehend = session.client(service_name='comprehend')
```

```python
ms=pd.read_csv('/content/dummy_dataset.csv')
```

```python
ms.head()
```

```python
def getSentiment(sentence):
    detected_sentiment=comprehend.detect_sentiment(Text=sentence, LanguageCode='en')['Sentiment']
    return detected_sentiment
```

```python
ms['sentiment_detected']=ms['content'].apply(getSentiment)
```
