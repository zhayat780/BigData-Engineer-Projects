```python
import pandas as pd 
import cloudy_warehouses
```

```python
import boto3
```

```python
service_name='s3'
region_name='us-east-2'
aws_access_key_id=''
aws_secret_access_key=''
```

```python
s3 = boto3.resource(
    service_name=service_name,
    region_name=region_name,
    aws_access_key_id=aws_access_key_id,
    aws_secret_access_key=aws_secret_access_key
)
```

```python
for obj in s3.Bucket('krishsirfollow').objects.all():
    df=pd.read_csv(obj.get()['Body'],index_col=None)
    df.columns=['SEPALLENGTH','SEPALWIDTH','PETALLENGTH','PETALWIDTH','CLASSNAME'];
    df.cloudy_warehouses.write_snowflake(table="video")
    print(df)
```
